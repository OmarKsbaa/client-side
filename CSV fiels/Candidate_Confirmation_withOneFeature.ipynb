```python
import pandas as pd

# Load datasets
koi_df = pd.read_csv(r"C:\Users\muhap\Downloads\cumulative_2025.10.03_07.53.41.csv")
confirmed_df = pd.read_csv("confirmed_dataset_clean.csv")

def match_by_period(koi_period, tolerance=0.001):
    """Return True if a confirmed exoplanet matches koi_period within tolerance."""
    matches = confirmed_df[
        (confirmed_df["pl_orbper"].notna()) &
        (confirmed_df["pl_orbper"].between(koi_period - tolerance, koi_period + tolerance))
    ]
    return not matches.empty
```

```python
# Only keep rows that have a koi_period
koi_df = koi_df[koi_df["koi_period"].notna()].copy()

# Apply matching function
koi_df["predicted_confirmed"] = koi_df["koi_period"].apply(lambda x: match_by_period(x))

# Ground truth: koi_disposition
koi_df["true_confirmed"] = koi_df["koi_disposition"].apply(lambda d: d.upper() == "CONFIRMED")
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

y_true = koi_df["true_confirmed"]
y_pred = koi_df["predicted_confirmed"]

print("Accuracy:", accuracy_score(y_true, y_pred))
print("\nClassification Report:\n", classification_report(y_true, y_pred))
print("\nConfusion Matrix:\n", confusion_matrix(y_true, y_pred))
```
